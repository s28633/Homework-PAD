Sformułowanie hipotezy H0 i HA

H0 nie ma istotnej różnicy w Ocenie końcowej pomiędzy uczniami i uczennicami.

HA: istnieje istotna różnica (bez wskazania kierunku) w Ocenie końcowej pomiędzy uczniami i uczennicami.

```python
df_F=data[data['plec'] =='F']
df_F
```

```python
df_M=data[data['plec'] =='M']
df_M
```

```python
df_F['ocena_koncowa'].describe()
```

```python
df_M['ocena_koncowa'].describe()
```

```python
t, p = stats.ttest_ind(df_M["ocena_koncowa"],df_F["ocena_koncowa"])
t, p
```

```python

if p < 0.05:
    print("Odrzucamy hipotezę zerową. Istnieje istotna różnica w ocenie końcowej pomiędzy uczniami i uczennicami.")
else:
    print("Nie ma podstaw, by odrzucić hipotezę zerową. Nie ma istotnej różnicy w ocenie końcowej pomiędzy uczniami i uczennicami.")
```

```python
import scipy.stats as stats 
import numpy as np
import pandas as pd
import plotly.graph_objects as go

scores= pd.read_csv("/content/drive/MyDrive/PJATK/PAD/ćwiczenia/PAD_07/spendingscores.csv")
scores
```

1- Male 0-Female

```python
scores['Gender'] =  scores.apply(lambda x: 1 if x['Gender'] == 'Male' else 0 if x['Gender'] == 'Female' else None, axis = 1)
```

```python
set(scores['Gender'])
```

```python
import statsmodels.formula.api as smf

model = smf.ols(formula="SpendingScore ~ Gender + Age + AnnualIncome", data=scores).fit()
print(model.summary())
```
